```python
try:
  x = int(input("Enter first number: "))
  y = int(input("Enter second number: "))
  result = x + y
  print(f"Addition: {x} + {y} = {result}")
except ValueError:
  print("Please enter valid Integers!")
```

```python
with open("single_input.txt", "w") as f:
  f.write("42")

try:
  with open("single_input.txt", "r") as f:
    num = int(f.read().strip())
    result = num * 2
    print(f"Single Input Read from File: {num} * 2 = {result}")
except FileNotFoundError:
  print("File not found!")
except ValueError:
  print("Error in reading or converting the input!")
```

```python
with open("multiple_input.txt", "w") as f:
  f.write("3\n5\n5")

try:
  with open("multiple_input.txt", "r") as f:
    numbers = [int(line.strip()) for line in f.readlines()]
    result = sum(numbers)
    print(f"Multiple Input Read from File: {numbers}, sum = {result}")
except FileNotFoundError:
  print("File not found!")
except ValueError:
  print("Error in reading or converting the input!")
```
